```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/content/student_data_250.csv")
print(df.columns)
```

```python
def plot_histogram(data, column, bins=10):
    """Plots a histogram for a given column."""
    plt.figure(figsize=(8, 5))
    sns.histplot(data[column], bins=bins, kde=True, color='skyblue')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def plot_scatter(data, x_col, y_col):
    """Plots a scatter plot with a trendline."""
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=data[x_col], y=data[y_col], color='red', alpha=0.7)
    m, b = np.polyfit(data[x_col], data[y_col], 1)
    plt.plot(data[x_col], m * data[x_col] + b, color='blue', linestyle="--")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'Scatter Plot of {x_col} vs {y_col}')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
def plot_bar_chart(data, x_col):
    """Plots a bar chart for categorical data."""
    plt.figure(figsize=(10, 5))
    major_counts = data[x_col].value_counts()
    major_counts.plot(kind='bar', color='green')
    plt.xlabel(x_col)
    plt.ylabel("Number of Students")
    plt.title(f'Bar Chart of {x_col}')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def plot_grouped_bar_chart(data, x_col, y_col):
    """Plots a grouped bar chart comparing students with/without extracurriculars by graduation year."""
    grouped_data = data.groupby([x_col, y_col]).size().unstack()
    grouped_data.plot(kind='bar', figsize=(10, 6), color=['blue', 'orange'])
    plt.xlabel(x_col)
    plt.ylabel("Number of Students")
    plt.title(f'Grouped Bar Chart of {y_col} Across {x_col}')
    plt.xticks(rotation=45)
    plt.legend(title=y_col)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
plot_histogram(df, 'GPA')
```

```python
plot_scatter(df, 'Attendance Rate (%)', 'GPA')
```

```python
plot_bar_chart(df, 'Major')
```

```python
plot_grouped_bar_chart(df, 'Graduation Year', 'Extracurricular Activities')
```
