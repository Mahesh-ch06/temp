# **Question**

The travel time (in minutes) per day for a person to travel from KU cross to SR Univeersity are recorded as follows

26,15, 22,15, 18, 15, 32, 15, 22, 15, 18, 23,24,18, 24,18, 26,18,22, 26,22, 23, 23, 32,24,25, 15, 25,26, 32, 36,48, 24

a) For how many days, the above data are reported ?

b) Find the sum of travel times for all those days

c) Construct a discrete distribution table

d) Add a relative frequency, percentage column in the dataframe

### **Reading Data**

```python
import pandas as pd

data = [26,15,22,15,18,15,32,15,22,15,18,23,24,18,24,18,26,18,22,26,22,23,23,32,24,25,15,25,26,32,36,48,24]
```

### **a) For how many days, the above data are reported ?**

```python
print("Length Of the Data")
num_days = len(data)
print(f"The data is reported for {num_days} days.")
```

### **b) Find the sum of travel times for all those days**

```python
print("Sum of Travel Times")
sum_travel_time = sum(data)
print(f"The total sum of travel times is {sum_travel_time} minutes.")
```

### **c) Construct a discrete distribution table**

```python
print("Value Count")
pd.Series(data).value_counts()
```

```python
df = pd.DataFrame(data, columns=['Travel_Time'])
distribution = df['Travel_Time'].value_counts().reset_index()
distribution.columns = ['Travel_Time', 'Frequency']
distribution = distribution.sort_values('Travel_Time').reset_index(drop=True)
distribution
```

```python
bins = [10, 20, 30, 40, 50]
labels = ['11-20', '21-30', '31-40', '41-50']
categorized_data = pd.cut(data, bins=bins, labels=labels, include_lowest=True)
frequency_table = categorized_data.value_counts().sort_index()
table_df = pd.DataFrame({'Range': frequency_table.index, 'Frequency': frequency_table.values})
table_df
```

### **d) Add a relative frequency, percentage column in the dataframe**

```python
distribution['Relative_Frequency'] = distribution['Frequency'] / len(data)
distribution['Percentage'] = distribution['Relative_Frequency'] * 100

distribution
```
